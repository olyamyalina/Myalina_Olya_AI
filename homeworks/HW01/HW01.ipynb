```python
import numpy as np
```

```python
arr_list = np.array([52, 67, 76, 44, 322])
print("Массив из списка:", arr_list)
```

```python
arr_zeros = np.zeros((2, 2))
print("Массив из нулей:\n", arr_zeros)
```

```python
arr_ones = np.ones((3, 3))
print("Массив из единиц:\n", arr_ones)
```

```python
arr_arange = np.arange(0, 10, 2)
print("Массив с помощью arange:", arr_arange)
```

```python
print("Форма массива arr_ones:", arr_ones.shape)
print("Тип данных массива arr_ones:", arr_ones.dtype)
```

```python
element = arr_list[1] 
print("Второй элемент массива arr_list:", element)
```

```python
sub_array = arr_list[0:2]
print("Подмассив arr_list[0:2] (0 и 1 элементы):", sub_array)
```

```python
arr_sum = arr_list + arr_arange
print("Сумма двух массивов arr_list и arr_arange: ", arr_sum)
```

```python
arr_sum_5 = arr_list + 5
print("Поэлементное сложение arr_list и 5: ", arr_sum_5)
```

```python
arr_prod = arr_list * arr_arange
print("Поэлементное умножение двух массивов arr_list и arr_arange:", arr_prod)
```

```python
arr_prod_3 = arr_list * 3
print("Поэлементное умножение массива arr_list и 3:", arr_prod_3)
```

```python
mean_value = arr_list.mean()
print("Среднее значение массива arr_list:", mean_value)
```

```python
sum_value = arr_list.sum()
print("Сумма элементов массива arr_list:", sum_value)
```
